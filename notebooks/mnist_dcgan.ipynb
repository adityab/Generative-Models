```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import gridspec
import sys
sys.path.append('..')
```

```python
from models.mnist_dcgan import MNIST_DCGAN

model = MNIST_DCGAN(z_dim=100)
model.load_saved('../mnist_dcgan-z100-epoch20.h5')

def sample_normal(model, n_samples):
    # Extract samples from the model
    latents = np.random.normal(0, 1, size=(n_samples, 100))
    return model.generate(latents)

def show_images(path, samples, rows, cols):
    # Save sampled images in a grid
    f = plt.figure(figsize=(8, 8))
    gs = gridspec.GridSpec(rows, cols)
    gs.update(wspace=0.0, hspace=0.0)
    k = 0
    for i in range(rows):
        for j in range(cols):
            img = samples[k]
            ax = plt.subplot(gs[k])
            ax.imshow(img, cmap='gray')
            ax.axis('off')
            k = k + 1

    #f.savefig(path, dpi=100, bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
# Sample 25 random images with interpolation
x = np.zeros((1, 100))
y = np.ones((1, 100))

for i in np.arange(0, 1, 0.05):
    plt.figure()
    plt.imshow(model.generate((1 - i) * x + i * y)[0], cmap='gray')
    
```
